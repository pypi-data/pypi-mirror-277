```python
import pandas as pd
df = pd.read_csv("fifa17-top20-women-fixed.txt",sep="-",header=None)
```

```python
rdf = df.rename(columns={0: "Name", 1: "Country", 2: "Rating"})
```

```python
countries = rdf.groupby("Country").count()
```

```python
rdf[rdf["Country"] == "USA"]
```

```python
cdf = rdf.assign(Name=rdf["Name"].str.strip(), 
                 Country=rdf["Country"].str.strip())
```

```python
cdf[cdf["Country"] == "USA"]
```

```python
country_counts = cdf.groupby("Country").size().sort_values(ascending=False)
```
