```python
import sys
sys.path.append("src")
import statfinpy
```

```python
db = statfinpy.PxWebAPI.Verohallinto()
tulot = db.table("Vero", "tulot_101.px")
```

```python
tulot.values["Tuloluokka"]
```

```python
df = tulot.query({
    "Verovuosi": 2022,
    "Erä": "HVT_TULOT_70",
    "Tulonsaajaryhmä": "Y",
    "Tuloluokka": "*",
    "Tunnusluvut": "*"
})
```

```python
tuloluokat = [str(x) for x in range(19,28)]
df = df[df.Tuloluokka.isin(tuloluokat)]
N = df.N.to_numpy().sum()
s = df.Sum.to_numpy().sum()
r = 88_200 * N
print(f"{int(N)}: {s/1e9:.2f} - {r/1e9:.2f} = {(s-r)/1e9:.2f} mrd.")
```
